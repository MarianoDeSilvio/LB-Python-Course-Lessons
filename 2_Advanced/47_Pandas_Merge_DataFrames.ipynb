```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# EXAMPLE FOR MERGE

job_data = {
    'job_id': [1, 2, 3, 4, 5],
    'job_title': ['Data Scientist', 'Software Engineer', 'Product Manager', 'Marketing Director', 'HR Manager'],
    'company_name': ['Google', 'Microsoft', 'Apple', 'Nike', 'Starbucks'],
    'job_location': ['Mountain View, CA', 'Redmond, WA', 'Cupertino, CA', 'Beaverton, OR', 'Seattle, WA']
}

company_data = {
    'company_name': ['Google', 'Microsoft', 'Apple', 'Nike', 'Starbucks'],
    'industry': ['Technology', 'Technology', 'Technology', 'Apparel', 'Food & Beverage'],
    'company_size': ['100,000+', '100,000+', '100,000+', '75,000+', '346,000+']
}


df_jobs = pd.DataFrame(job_data)
df_companies = pd.DataFrame(company_data)
```

```python
print('df_jobs')
print(df_jobs)
print('')
print('df_companies')
print(df_companies)
```

```python
df_jobs.merge(df_companies, on='company_name')
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B')
df_US_pivot['month_no'] = df_US_pivot['month_no'].dt.month
df_US_pivot.sort_values(by = 'month_no', inplace = True)
df_US_pivot.set_index('job_posted_month', inplace = True)
df_US_pivot = df_US_pivot.drop(columns='month_no')
df_US_pivot
```

```python
pd.read_csv('https://lukeb.co/software_csv')
```

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')
```

```python
df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
top_5 = df_US_merged.sum().sort_values(ascending = False).head()
top_5
```

```python
top_5 = top_5.index.to_list()
top_5
```

```python
df_US_merged[top_5].plot(kind='line')
plt.show()
```

```python
# CLEANED UP PLOT

custom_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df_US_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20_000)
plt.xticks(ticks=range(len(custom_labels)), labels=custom_labels)
plt.legend()
plt.show()
```
